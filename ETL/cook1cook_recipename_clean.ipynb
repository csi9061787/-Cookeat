```python
from pymongo import MongoClient
import pandas as pd
import re
from difflib import *
```

```python
client = MongoClient('127.0.0.1', 27017)
collection = client["food"]["cook1cook"]
data = collection.find()
df = pd.DataFrame(data)
df.head()
```

```python
recipe_name = df['RecipeName']
recipe_name
```
